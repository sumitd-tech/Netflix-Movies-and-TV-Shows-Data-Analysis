```python
# Drop exact duplicate rows
df.drop_duplicates(inplace=True)

# Normalize column names: lower, strip, replace spaces with underscore
df.columns = [c.strip().lower().replace(' ', '_') for c in df.columns]
print("Normalized columns:", df.columns.tolist())

# Trim whitespace in object columns
for c in df.select_dtypes(include='object').columns:
    df[c] = df[c].astype(str).str.strip().replace({'nan': np.nan})
```

```python
# map possible names to standard names we'll use
col_map = {}
c = df.columns

# title
if 'title' in c:
    col_map['title'] = 'title'
elif 'name' in c:
    col_map['title'] = 'name'

# type / genre
if 'type' in c:
    col_map['type'] = 'type'
elif 'genres' in c:
    col_map['type'] = 'genres'
elif 'listed_in' in c:
    col_map['type'] = 'listed_in'   # common on some Netflix datasets

# country
if 'country' in c:
    col_map['country'] = 'country'

# release year / date
if 'release_year' in c:
    col_map['release_year'] = 'release_year'
elif 'release_date' in c:
    col_map['release_year'] = 'release_date'  # we'll parse to year later
elif 'date_added' in c:
    col_map['release_year'] = 'date_added'

# rating, duration, director, cast (optional)
for opt in ['rating','duration','director','cast']:
    if opt in c:
        col_map[opt] = opt

print("Detected column mapping (existing -> used):")
display(col_map)
```

```python
# Replace separators and convert to list
if 'type' in df.columns:
    # typical separators: ', ', ',','|'
    df['type'] = df['type'].astype(str).replace({'nan': np.nan})
    df['type_list'] = (df['type']
                       .fillna('')
                       .str.replace(r'\s*\|\s*', ',', regex=True)
                       .str.split(',')
                       .apply(lambda lst: [s.strip() for s in lst if s and s.strip() != '']))
else:
    df['type_list'] = [[]] * len(df)
```

```python
# Decide what is critical: title and release_year and type_list
critical = []
if 'title' in df.columns: critical.append('title')
if 'release_year' in df.columns: critical.append('release_year')

print("Dropping rows missing:", critical)
df.dropna(subset=critical, inplace=True)
print("After drop shape:", df.shape)
```

```python
clean_path = '/content/Netflix Dataset (8).csv'
df.to_csv(clean_path, index=False)
print("Saved cleaned CSV to:", clean_path)
```

```python
# Debug: show columns so you can confirm names
print("Columns in df:", df.columns.tolist())

# 1) Create/ensure release_year exists from common date columns
date_candidates = ['release_year', 'release year', 'date_added', 'release_date', 'year', 'date']
created = False

for col in date_candidates:
    if col in df.columns:
        print(f"Using column '{col}' to create 'release_year'.")
        # Try datetime parse -> year
        try:
            df['release_year'] = pd.to_datetime(df[col], errors='coerce').dt.year
            # If parse failed (all NaN), try numeric conversion
            if df['release_year'].isna().all():
                df['release_year'] = pd.to_numeric(df[col], errors='coerce')
        except Exception as e:
            # fallback numeric
            df['release_year'] = pd.to_numeric(df[col], errors='coerce')
        created = True
        break

if not created:
    # last resort: if you have a numeric 'release' style column or nothing at all
    if 'release' in df.columns:
        df['release_year'] = pd.to_numeric(df['release'], errors='coerce')
        created = True

if not created:
    raise KeyError("No date-like column found to create 'release_year'. Check df.columns and provide a suitable column.")

# Convert to integer type (nullable)
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce').astype('Int64')

# Optional: inspect distribution of release_years
print("release_year value counts (sample):")
print(df['release_year'].value_counts(dropna=True).head(10))

# 2) Ensure media_type exists (Movie vs TV Show)
# Use an existing 'type'/'listed_in'/'genres' column if present
type_candidates = ['type', 'listed_in', 'genres', 'category']
type_col = None
for col in type_candidates:
    if col in df.columns:
        type_col = col
        break

# create a normalized text column to inspect
if type_col:
    df['type_norm'] = df[type_col].astype(str).str.lower().str.strip()
else:
    df['type_norm'] = ''

# Map common keywords to 'Movie' / 'TV Show'
df['media_type'] = None
df.loc[df['type_norm'].str.contains('movie', na=False), 'media_type'] = 'Movie'
df.loc[df['type_norm'].str.contains('film', na=False), 'media_type'] = 'Movie'
df.loc[df['type_norm'].str.contains('tv', na=False) | df['type_norm'].str.contains('show', na=False), 'media_type'] = 'TV Show'

# If still missing, try a fallback using a 'duration' column: many Netflix datasets store durations like "90 min" or "1 Season"
if df['media_type'].isna().all() and 'duration' in df.columns:
    dur = df['duration'].astype(str).str.lower()
    df.loc[dur.str.contains('season', na=False) | dur.str.contains('seasons', na=False), 'media_type'] = 'TV Show'
    df.loc[dur.str.contains('min', na=False) | dur.str.contains('minutes', na=False), 'media_type'] = 'Movie'

# Fill remaining unknowns with 'Unknown' (so groupby won't fail)
df['media_type'] = df['media_type'].fillna('Unknown')

# 3) Prepare for grouping: drop rows without release_year if you want only known years
print("Rows before dropping missing release_year:", len(df))
df_plot = df.dropna(subset=['release_year']).copy()
print("Rows with release_year available:", len(df_plot))

# 4) Group and pivot safely
media_by_year = (df_plot
                 .groupby(['release_year', 'media_type'])
                 .size()
                 .unstack(fill_value=0)
                 .sort_index())

print("media_by_year sample:")
print(media_by_year.head())

# 5) Plot (matplotlib stacked area)
ax = media_by_year.plot(kind='area', stacked=True, figsize=(12,5),
                        title='Movies vs TV Shows by Release Year')
ax.set_xlabel('Release Year')
ax.set_ylabel('Count')
ax.legend(title='Media Type')
plt.show()

# 6) Plotly interactive area (reset index for plotly)
fig = px.area(media_by_year.reset_index(), x='release_year',
              y=media_by_year.columns.tolist(),
              title='Movies vs TV Shows by Release Year (interactive)')
fig.show()
```

```python
# Explode the type_list so each row = one title + one genre
df_expl = df.explode('type_list').rename(columns={'type_list':'genre'})
df_expl['genre'] = df_expl['genre'].astype(str).str.strip().replace({'nan': np.nan})
df_expl = df_expl[~df_expl['genre'].isna() & (df_expl['genre'] != '')]

# Top genres overall
top_genres = df_expl['genre'].value_counts().head(15)
print("Top genres:\n", top_genres)

# Genre counts over time (example: top 6 genres)
top6 = top_genres.head(6).index.tolist()
genre_trends = (df_expl[df_expl['genre'].isin(top6)]
                .groupby(['release_year','genre']).size()
                .unstack(fill_value=0).sort_index())

# Plot trends
genre_trends.plot(figsize=(12,6), title='Top 6 Genres Over Years')
plt.xlabel('Release Year'); plt.ylabel('Count'); plt.legend(title='Genre')
plt.show()
```

```python
# Clean country column if detected
if 'country' in df.columns:
    df['country_clean'] = df['country'].astype(str).str.split(',').str[0].str.strip()
    top_countries = df['country_clean'].value_counts().head(12)
    print("Top countries:\n", top_countries)

    # Bar plot
    plt.figure(figsize=(10,5))
    sns.barplot(x=top_countries.values, y=top_countries.index)
    plt.title('Top Countries by Number of Titles')
    plt.xlabel('Number of Titles'); plt.ylabel('Country')
    plt.show()

    # Country share pie (plotly)
    fig = px.pie(values=top_countries.values, names=top_countries.index, title='Top Countries Share')
    fig.show()
else:
    print("No country column detected.")
```
